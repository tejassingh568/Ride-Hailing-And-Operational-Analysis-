```python
import pandas as pd 
```

```python
df = pd.read_csv(r"C:\Users\My Pc\Downloads\bangalore_ride_dataset_100k_4STATUS_FIXED.csv")
```

```python
df.head()

```

```python
df["Time"] = df["Time"].astype(str).str.strip()
df["Time"] = pd.to_datetime(df["Time"], errors="coerce").dt.time
df.head()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace(r"[^\w]", "_", regex=True)
df.columns
```

```python
import numpy as np
df = df.replace({"NA": np.nan, "": np.nan})

```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df["order_value_slab"] = pd.cut(
    df["booking_value"],
    bins=[0, 500, 1000, 10000],
    labels=["Low", "Medium", "High"],
    right=False
)
```

```python
df.head()
```

```python
median = df.loc[df["booking_status"] == "Success", "customer_rating"].median()
df.loc[(df["booking_status"] == "Success") & df["customer_rating"].isna(),
       "customer_rating"] = median
```

```python
df.isnull().sum()
```

```python
driver_median = df.loc[df["booking_status"] == "Success", "driver_ratings"].median()
df.loc[
    (df["booking_status"] == "Success") & df["driver_ratings"].isna(),
    "driver_ratings"
] = driver_median
```
